```python
#https://github.com/djw-easy/GeoDetector/blob/main/example.ipynb
#https://pypi.org/project/geodetector/
#GUNAKAN RSEI DATA Y DAN X YANG DIINTERSECT DALAM FORMAT PIKSEL KE POINT
#INPUT DATA MASUKKAN DATA FORMAT .XLS SHEET NAME BERUPA NAMA SHEET NAME NYA
#Data
#Y = RSEI Trend 2023 (piksel)
#X1 = Kemiringan Lereng (piksel)
#X2 = Ketinggian    (piksel)
#X3 = Jarak Dari Jalan (piksel)
#X4 = Jarak Dari Sungai (piksel)
#X5 = Jarak Dari Pemerintahan  (piksel) 
#X6 = Jarak Dari Pendidikan (piksel)
#X7 = Jarak Dari Kesehatan  (piksel)
#X8 = Curah Hujan   (piksel)
#X9 = Kepadatan Penduduk    (piksel)
#x10 = Pendapatan Perkapita (piksel)

import pandas as pd
from geodetector import (
    factor_dector, 
    interaction_detector, 
    ecological_detector
)
```

```python
df = pd.read_excel('trial.xlsx',
                   sheet_name='trial')
df.head()
```

```python
factor_dector(df, 'Y', [ 'X1','X2','X3', 'X4','X5', 'X6', 'X7', 'X8', 'X9', 'X10'])
```

```python
interaction_detector(df, 'Y', ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10']) 
```

```python
df1, df2 = interaction_detector(df, 'Y', ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10'], relationship=True)
df1
```

```python
ecological_detector(df, 'Y', ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10'])
```
